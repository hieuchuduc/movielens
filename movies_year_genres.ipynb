```python
movies = pd.read_csv('./movielens/movies_all.csv')
```

```python
movies = movies.drop('Unnamed: 0', axis=1)
```

```python
title_new = list()
year = list()
```

```python
for index, row in movies.iterrows():
    title = row['title']
    pat = 0
    for i in range(len(title)-1):
        if ( (title[i]=='(') & (title[i+1].isdigit()) ):
            pat = i
            break
    year.append(row['title'][pat+1:])
    title_new.append(row.loc['title'][:pat-1])
```

```python
title_new = pd.Series(title_new)
year = pd.Series(year)
title_new
```

```python
movies = pd.concat(axis=1,objs=[movies, title_new, year])
```

```python
movies.tail(30)
```

```python
movies = movies.drop('title', axis=1)
```

```python
movies = movies.rename(columns={0:'title', 1:'year'})
```

```python
order = ['movieId','title','year','genres','n_users','rating_mean','rating_std']
movies = movies[order]
```

```python
movies['year'] = movies['year'].apply(lambda x: x[:len(x)-1])
```

```python
movies.head(15)
```

```python
movies.to_csv('./movielens/movies_all.csv')
```

```python
movies_new = pd.DataFrame.copy(movies)
movies_new.tail(10)
```

```python
movies_new.genres = movies_new.genres.str.split(pat='|')
```

```python
movies_new.head(10)
```

```python
movies.head(10)
```

```python
movies_genres = dict()
```

```python
for index, row in movies_new.iterrows():
    for genre in row['genres']:
        if (not(genre in movies_genres)):
            movies_genres[genre] = []
        movies_genres[genre].append(index)
```

```python
for key, value in movies_genres.items():
    movies.iloc[value].to_csv(key + '.csv')
```
